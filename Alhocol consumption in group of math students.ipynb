```python
notAtHome=df1.drop(index='at_home')
```

```python
x=notAtHome.value_counts('romantic')
x
```

```python
countOfRomantic=x.no+x.yes
countOfRomantic
```

```python
percentageOfYes=x.yes/countOfRomantic
percentageOfYes
```

```python
percentageOfNo=x.no/countOfRomantic
percentageOfNo
```

```python
x=atHome.value_counts('romantic')
countOfRomanticAtHome=x.no+x.yes
```

```python
percentageOfNoAtHome=x.no/countOfRomanticAtHome
percentageOfYesAtHome=x.yes/countOfRomanticAtHome
```

```python
d={'Mother Working':['working','working','not working','not working'],'percentage':[percentageOfNo,percentageOfYes,percentageOfNoAtHome,percentageOfYesAtHome],'In romantic relation':['No','Yes','No','Yes']}
d=pd.DataFrame(data=d)
```

```python
d
```

```python
fig= px.bar(data_frame=d,x='Mother Working',y='percentage',color='In romantic relation',barmode='group',width=500)
fig.show()
```
